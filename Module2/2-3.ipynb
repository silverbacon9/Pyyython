|基本型態|創建方式|
|:---:|:---:|
| int | `num = 10`  |
| float | `num = 2.5`  |
| boolean |  `parse_ok = True` |
| string | `name = "Luke"` |

|容器型態|創建方式|
|:---:|:---:|
|list| `num_list = [1,2,3]` | |
|tuple| `num_tuple = (1,2,3)` |
|set| `num_set = {1,2,3,4,5}`|
|dict| `user_info = {"name":"Luke,"age"="28"}`|


整數 (int)、浮點數 (float)

```python
num1 = 10
num2 = 2.5
print(num1)
print(num2)
```

布林值 (boolean)

```python
parse_ok = True
parse_no = False
print(parse_ok)
print(parse_no)
```

字串 (string)

```python
message = "hello world!"
print(message)
```

```python
data = "True"
parse_ok = True
print(data)
print(parse_ok)

# type()
print(type(data))
print(type(parse_ok))
```

列表 (list)

```python
num_list = [1, 2, 3, 4, 1]
print(num_list)
print(type(num_list))
```

元組 (tuple)

```python
num_tuple = (1, 2, 3, 2)
print(num_tuple)
print(type(num_tuple))
```

集合 (set)

```python
num_set = {1, 2, 3, 1, 5, 9, 7, 3,}
print(num_set)
print(type(num_set))
```

字典 (dict)

```python
user_info = {"name":"Luke","age":25}
print(user_info)
print(type(user_info))
```

```python
data_list = []
print(data_list)
print(type(data_list))
# 創建一個空的 dict

x = {}
print(x)
print(type(x))
```

```python
# 創建一個空的 set
empty_set = set()
print(empty_set)
print(type(empty_set))
```
