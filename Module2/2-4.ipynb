賦值運算子

```python
a = 3
b = 4

b += a # 等同於 b = b + a 
print(b)

b -= a # 等同於 b = b - a
print(b) #若變數名稱重複，則運算會以上一次計算得出的值再進行計算
```

```python
a = 3
b = 4
b *= a 
print(b)
```

比較運算子 (用 if 判斷)

```python
print(5 == 5)
a = 10
b = 5
print(a == b)
print(a != b)
```

```python
a = 10
b = 5

if a != b :
    print("A不等於B")
```

```python
score = 45
passing_score = 60
if score < passing_score :
    score += 10
    print("加分")
    print(score)   
# tab 縮排
# shift + tab 反縮排
print("學生的分數是:", score)
```

邏輯運算子 and / or / not

```python
name = "Alex"
age = 18
if name == "Alex" and age == 18 :
    print("name: Alex ; age: 18")
```

```python
name = "Alex"
age = 18
if name == "Alex" or age == 18 :
    print("正確")
```

```python
if not(name == "Luke"):
    print("執行 not 語句")
```

成員運算子 in / not in

```python
num_list = [1,2,3,4]
if 6 in num_list:
    print("1 in the list")
print(1 in num_list)
```

```python
if 5 not in num_list:
    print("5 not in the list")
```

身分運算子 is / is not

```python
x = 10
y = 10
print(x)
print(y)
print(id(x))
print(id(y))
```

```python
if x is y:  # 不要用is去判斷"值"是不是相等
    print("x 和 y參照到同一個記憶體位置")
```

```python
x = [1, 2, 3]
y = [1, 2, 3]
if x == y :
    print("x 等於 y")

print(id(x))
print(id(y))

if x is y :
    print("x 和 y 記憶體位置相同")
```
