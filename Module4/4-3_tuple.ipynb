```python
# tuple 初始化: 第一種
data = ("big", "data", "is", "Good")
print(type(data))
print(data)

```

```python
# tuple 初始化: 第二種
data = "Jarvis" , 18
print(type(data))
print(data)

name, age = "Jarvis", 18
print(name)
print(age)
```

```python
data = "Jarvis", 18, "AIPE02"
print(data)

name, age, _class = data
print(name)
print(age)
print(_class)
```

```python
# 透過指定索引輸出值
data = "Jarvis", 18, "AIPE02"
print(data)
print(data[0])
```

```python
# tuple 不可以修改指定索引的值
data[0] = 100
```

```python
# (list)可以修改指定索引的值
num_list = [1, 2, 3]
num_list[0] = 100
print(num_list)
```

```python
# 用 for 迴圈逐一輸出資料
data = "Jarvis", 18, "AIPE02"
for i in data:
    print(i)
    
```

```python
# 用 len 計算 tuple 資料個數
data = "Jarvis", 18, "AIPE02"
print(len(data))
# ↑實務上會先用len確定資料筆數或是資料長度，確認後才會用print確認資料內容
print(data)

```
