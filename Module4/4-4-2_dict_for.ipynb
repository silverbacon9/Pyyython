```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['greem', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': 'orange' # 錯誤的寫法
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1 
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2
# 取 dict 中的 value
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料3
for key, value in fruit_dict.items():
    print(key)
    print(value)
```

##### `defaultdict(<type>)`

```python
from collections import defaultdict
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
data = defaultdict(int)

# 以int型態的預設值 0 來初始化
print(data["price"])
print(data)
```
