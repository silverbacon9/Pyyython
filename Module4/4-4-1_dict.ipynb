```python
# 初始化 dict
# 若要用 dict ，變數命名要夠清楚，變數名稱長一點也沒關係
fruit_price_dict = {"apple": 100, "orange": 50, "tomato": 30}
print(fruit_price_dict)
```

```python
# 印出 蘋果的價格
# 選dict 中變數的 key，可印出相對應的 value 
fruit_price_dict = {"apple": 100, "orange": 50, "tomato": 30}
print(fruit_price_dict["apple"])
```

```python
# 修改 橘子的價格
fruit_price_dict = {"apple": 100, "orange": 50, "tomato": 30}
fruit_price_dict["orange"] = 70
print(fruit_price_dict)
```

```python
# 刪除 番茄
fruit_price_dict = {"apple": 100, "orange": 50, "tomato": 30}
del fruit_price_dict["tomato"]
print(fruit_price_dict)
```

```python
fruit_price_dict = {"apple": 100, "orange": 50, "tomato": 30}
fruit_price_dict.pop("tomato")
print(fruit_price_dict)
```

```python
fruit_price_dict = {
        "apple": 100,
        "orange": 50,
        "tomato": 30
    }
print(fruit_price_dict)

# dict 項目多的時候的建議寫法
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name": "Jarvis",
    "age": 20,
    "info":{
        "nickname": "Lu",
        "favorite_role": "ironman",
        "phone": [
            "0912-345-678",
            "0923-456-789",
        ]
    }

}
```

```python
# 取得 nickname
print(user_info["info"])
print(user_info["info"]["nickname"])
```

```python
# 取得所有 phone_number
print(user_info["info"]["phone"])
for phone_number in user_info["info"]["phone"]:
    print(phone_number)
```

##### dict 的合併

```python
# dict 的合併
fruit_dict2= {
    "kiwi":{
        "price":30,
        "amount": 20,
        "color": ["green"]
    }
}
fruit_dict.update(fruit_dict2)
print(fruit_dict)
```

```python
print(fruit_dict)
fruit_dict3= {
    #兩個米字號，代表把dict做拆解，取得全部的value，再新作成一個新的 dict
    **fruit_dict,
    "kiwi":{
        "price":30,
        "amount": 20,
        "color": ["yellow"]
    },
}
print(fruit_dict3)
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
# 打印出來的資料沒有按照順序
from pprint import pprint
pprint(fruit_dict3)
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''

fruit_dict3["apple"]["color"].append("yellow")
fruit_dict3["tomato"]["color"].append("green")
pprint(fruit_dict3)
```
