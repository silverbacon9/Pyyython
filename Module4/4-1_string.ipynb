|string方法| 功能 |
|:---:|:---:|
| `str.replace() | 替換字串 |
| `str.stripe() | 去除兩側空格 |
| `str.upper() | 字串小寫變大寫 |
| `str.lower() | 字串大寫變小寫 |

```python
# 字串變數初始化
name = "Jarvis"
print(name)
print(name[0])
```

```python
# 替換字串
# 方法 也是 function 的一種， function 不一定是 方法
# 定義在class底下的 function 可以稱為 method
'''
str.replace()
↑呼叫 function name 再加上 ()

將 string 中的 str1 替換成 str2
'''
name = "Alex"
new_name = name.replace("ex", "len")
# ↑要設定一個新的變數去接修改過後的值，因字串不可變
# ↑只要遇到ex，就替換成lens
print(new_name)
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''
data = " |__data__| "
print(data)

new_data = data.strip()
print(new_data)
```

```python
data = "111 |__data__|   1"
print(data)

new_data = data.rstrip("1")
# new_data = data.lstrip("1")
print(new_data)
```

```python
# 字串小寫變大寫
'''
str.lower()
將字串 string 裡的字母全部改成小寫
'''
data = "CAR"
new_data = data.lower(data)
print(new_data)

'''

將字串 string 裡的字母全部改成大寫
'''

data = "car"
new_data = data.upper(data)
print(new_data)
```

```python
# 字串分割成 list
# str.split()
greeting = "hello world! hello John"

str_list = greeting.split("!")
# split預設適用空格去隔開各字串.split(" ")
print(str_list)
```

```python
for data_str in str_list:
    print(data_str)
    
```
