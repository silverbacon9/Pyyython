##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
for i in 'content':
    print(i)
    if i == 't':
        break
print('迴圈結束')

```

```python
for i in range(1, 100):
    print(i)
    break
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
for char in 'content':
    if char == 't':
        continue
    print(char)
    
```

### 小試身手
計算 1 到 100 之間的和，如果值超過500就停止執行

```python
a = 0
for i in range(1, 101):
    a += i
    if a > 500:
        break
print(a)
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python
# total = 0
# for i in range(1, 101):
#     if i % 2 != 0 and i % 5 != 0:
#         print(i, end=' ')
#         total += i
# print()
# print(total)

total = 0
for i in range(1, 101):
    if i % 2 == 0 or i % 5 != 0:
        continue
    else:
        print(i, end=' ')
        total += i
print()
print(total)
```
