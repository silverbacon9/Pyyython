| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
# ctrl + D 選擇相同的文字做修改
# for <變數> in 可迭代的物件
n = 1
for char in "content":
    print(char, end=',')
    print(n, end=',')
    n += 1
```

```python
# range(m) 走訪 0 到 m-1 的數字
for i in range(5):
    print('hello world!')
```

```python
# range(n, m) 走訪 n 到 m-1 的數字
for i in range(1, 11):
    print(i, end=',')

```

```python
# range(n, m-1, step) 以 step 為間距，走訪 n 到 m-1 的數字 
for i in range(1, 11, 2):
    print(i)
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
a = 4
n = 5
for i in range(n):
   if a % 2 == 0:
      print(a+1+i*2)
   else:
      print(a + 1 + i*2)
```

```python
a = 4
n = 5
if a %2 != 0:
    a +=1
for i in range(n):
    print(a+ i*2 )
```

```python
n = 4
for a in range(n):
    print(a+2)
```

##### while

```python
count = 1
while count <= 5:
    print(count, end=',')
    count += 1
print('count 已經不小於5 結束迴圈')
print(count)

```
